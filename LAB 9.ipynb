> ## <font color='red'> 1. Generate normal random process and plot the probability density function (PDF) of random variable for different value of $t$. Find out whether the process is wide sense stationary (WSS).<font>



```python
time_size = 1000
samples_size = 1000 
tau = 10 
mu = 12 
sigma = 1 

sample_func = np.zeros((samples_size,time_size))

for i in range(samples_size):
    sample_func[i] = np.random.normal(mu,sigma,time_size)
# print(sample_func[4])

# pdf at t=10
t = 10
arr = []
for i in range(samples_size):
    arr.append(sample_func[i][t])
plt.figure()
plt.hist(arr)

# pdf at t=50
t = 50
arr = []
for i in range(samples_size):
    arr.append(sample_func[i][t])
plt.figure()
plt.hist(arr)

mn = 99999
mx = 0

for i in range(time_size):
    avg = np.mean(sample_func[:][i])
    mn = min(mn,avg)
    mx = max(mx,avg)
print(np.abs(mn-mx))

col1 = sample_func[:][10]
col2 = sample_func[:][10+tau]

col1 = col1*col2
print(np.mean(col1))


col1 = sample_func[:][20]
col2 = sample_func[:][20+tau]

col1 = col1*col2
print(np.mean(col1))

#write code here
```

## <font color='red'> 2. For the random process given by $X(t)=A \cos(\omega_ot+\theta)$, where $\theta$ is a random variable uniformly distributed between $0$ and $2\pi$. Find out whether the process is wide sense stationary (WSS).<font>

```python
time_size = 1000
samples_size = 10000
Amplitude = 1
fc = 10
theta = np.random.uniform(0,2*np.pi,samples_size)

t = np.arange(0,1,1/time_size)

sample_func = np.zeros((samples_size,time_size))
for i in range(samples_size):
    sample_func[i] = Amplitude * np.cos(2*np.pi*fc*t + theta[i])
    
plt.plot(sample_func[4])

mn = 99999
mx = 0

for i in range(time_size):
    avg = np.mean(sample_func[:][i])
    mn = min(mn,avg)
    mx = max(mx,avg)
print(np.abs(mn-mx))

col1 = sample_func[:][10]
col2 = sample_func[:][10+tau]

col1 = col1*col2
print(np.mean(col1))


col1 = sample_func[:][20]
col2 = sample_func[:][20+tau]

col1 = col1*col2
print(np.mean(col1))


#write code here
```

## <font color='red'> 3. Let $Z(t)=X\cos(\omega_ct)+Y\sin(\omega_ct)$, where $X$ and $Y$ are independent Gaussian random variables, each with 0 mean and unit variance. Show that $Z(t)$ is WSS and $R_z(\tau)=\cos(\omega_ct)$. Let $Z(t_1)=V$. Show that $V$ is $N(0,1)$.<font>

```python
mu = 0
sigma = 1
sample_size = 100000
time_size = 100
fc = 10
t = np.arange(0,1,1/time_size)
X = np.random.normal(mu,sigma,sample_size)
Y = np.random.normal(mu,sigma,sample_size)

sample_func = np.zeros((sample_size,time_size))
for i in range(sample_size):
    sample_func[i] = X[i]*np.cos(2*np.pi*fc*t) + Y[i]*np.sin(2*np.pi*fc*t) 
    
# checking if given functions are normal or not

t1 = 10
arr = []
for i in range(sample_size):
    arr.append(sample_func[i][t1])
plt.figure()
plt.hist(arr)

t1 = 20
arr = []
for i in range(sample_size):
    arr.append(sample_func[i][t1])
plt.figure()
plt.hist(arr)

# checking for wss process

mn = 99999
mx = 0

for i in range(time_size):
    avg = np.mean(sample_func[:][i])
    mn = min(mn,avg)
    mx = max(mx,avg)
print(np.abs(mn-mx))
tau = 5
col1 = sample_func[:,10]
col2 = sample_func[:,10+tau]

col1 = col1*col2
print(np.mean(col1))


col1 = sample_func[:,20]
col2 = sample_func[:,20+tau]

col1 = col1*col2
print(np.mean(col1))

arr = []
for i in range(time_size):
    tau = i
    col1 = sample_func[:,0]
    col2 = sample_func[:,0+tau]
    arr.append(np.mean((col1*col2)))
plt.figure()
plt.plot(arr)
```
